# Building ML Models on the data

```python
'''
We are going to build a model that trains the given data of diabetic and prediabetic patients 
for multi-output regression using
ML regression models:
--->LinearRegression 
--->KNeighborsRegressor
--->DecisionTreeRegressor
--->RandomForestRegressor 
Deep Learning Models for Multi-Output Regression
--->Neural Network for Multi-Output Regression

'''
```

```python
from numpy import absolute
from numpy import mean
from numpy import std
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

#cross validation

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold

```

```python
#creating training and testing datasets for the training and testing 
from sklearn.model_selection import train_test_split
# diabetes_training , diabetes_testing = train_test_split(diabetes_df,test_size = 0.2,random_state = 25)
# print(diabetes_training.shape,diabetes_testing.shape)
```

```python
inputs = ['A1C_Test_perCnt', 'Fasting_Blood_Sugar', 'Glucose_Tolerance_Test','Random_Blood_Sugar','Age','BMI']
results=['methi_seeds', 
'water', 
'Millet', 
'Apple',
'ButterMilk', 
'Grek Yoghurt',
'Fatty Fish',
'ragi',
'Jawar', 
'Sprouts'] 
x = diabetes_df[inputs]
y = diabetes_df[results]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state =25)
```

```python
x_train.shape,y_train.shape
```

##  Multi-output regression
## Linear Regression

```python
from sklearn.linear_model import LinearRegression

#define model
LR_model = LinearRegression()
#Cross Validation procedure preparation
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# evaluate the model and collect the scores
n_scores = cross_val_score(LR_model, x_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force the scores to be positive
n_scores = absolute(n_scores)
# summarize performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

## k-Nearest Neighbors for Multioutput Regression

```python
from sklearn.neighbors import KNeighborsRegressor
#define model
knn_model = KNeighborsRegressor()
#Cross Validation procedure preparation
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# evaluate the model and collect the scores
n_scores = cross_val_score(knn_model, x_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force the scores to be positive
n_scores = absolute(n_scores)
# summarize performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

## Random forest regressor

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
```

```python
from sklearn.ensemble import RandomForestRegressor
#define model
rf_model = RandomForestRegressor()
#Cross Validation procedure preparation
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# evaluate the model and collect the scores
n_scores = cross_val_score(rf_model, x_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force the scores to be positive
n_scores = absolute(n_scores)
# summarize performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

## Decision Tree for Multioutput Regression

```python
from sklearn.tree import DecisionTreeRegressor
#define model
DT_model = KNeighborsRegressor()
#Cross Validation procedure preparation
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# evaluate the model and collect the scores
n_scores = cross_val_score(DT_model, x_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force the scores to be positive
n_scores = absolute(n_scores)
# summarize performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

## Wrapper Multioutput Regression Algorithms
 ### Direct Multioutput,Chained Multioutput


```python
#Support vector regressor
from sklearn.svm import LinearSVR
#Direct multioutput regresson
from sklearn.multioutput import MultiOutputRegressor
# define base model
SVR_model = LinearSVR()
# define the direct multioutput wrapper model
wrapper = MultiOutputRegressor(SVR_model)
# evaluate the model and collect the scores
n_scores = cross_val_score(wrapper, x_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force the scores to be positive
n_scores = absolute(n_scores)
# summarize performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
#Support vector regressor
from sklearn.ensemble import RandomForestRegressor
#Direct multioutput regresson
from sklearn.multioutput import MultiOutputRegressor
# define base model
rf_model = RandomForestRegressor()
# define the direct multioutput wrapper model
wrapper = MultiOutputRegressor(SVR_model)
# evaluate the model and collect the scores
n_scores = cross_val_score(wrapper, x_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force the scores to be positive
n_scores = absolute(n_scores)
# summarize performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

## Chained Multioutput Regression

```python
from sklearn.multioutput import RegressorChain
from sklearn.svm import LinearSVR

# define base model
SVR_model = LinearSVR()
# define the direct multioutput wrapper model
wrapper_chain = RegressorChain(SVR_model)
# evaluate the model and collect the scores
n_scores = cross_val_score(wrapper_chain, x_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force the scores to be positive
n_scores = absolute(n_scores)
# summarize performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```
